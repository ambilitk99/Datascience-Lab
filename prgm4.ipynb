```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,4,6])
y=np.array([1,2,4,6])
plt.plot(x,y,marker='*',linestyle='--',color='r')
plt.show()
```

```python
y=np.array([35,25,25,15])
a=["Apple","Green","Yellow","Blue"]
c=["red","blue","green","yellow"]
d=[0.3,0,0,0]
plt.pie(y,labels=a,colors=c,explode=d,shadow=True)
plt.show()
```
