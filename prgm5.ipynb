```python
# Rotation

from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
f1=Image.open('f1.png')
b=np.array(f1)
c=ndimage.rotate(b,270,cval=110)
d=Image.fromarray(c)
d.save("f1.png")
plt.imshow(d)
```

```python
# Scaling

rom PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
f1=Image.open('f1.png')
b=np.array(f1)
c=ndimage.zoom(b,(.25,1,1))
d=Image.fromarray(c)
d.save("f1.png")
plt.imshow(d)
```
